# Data Dictionary

|  Column Name        |  Field Name              |  Data Type  |  Description  |
|---------------------|--------------------------|-------------|------------------------------------|
|team_name            |NFL Team Name             |String       |The name of the NFL team this record refers to
|season_year          |Season                    |Integer      |The calendar year when the NFL season begins
|week                 |NFL Week                  |Integer      |The week (Thursday-Wednesday) of the NFL season
|day                  |Day of the Week           |String       |The day of the week the NFL game was played on
|date                 |Date                      |Date         |The calendar date of the NFL game
|game_time            |Game Time                 |String       |The time the game starts in Eastern Time
|rest_days            |Rest Days                 |Integer      |The number of days since the team's previous game
|home_team            |Is Home Team?             |Boolean      |Is the team hosting the game or not
|stadium_name         |Stadium Name              |String       |Where the game was played
|location             |Location                  |String       |What city and state/country the game was in
|latitude             |Latitude                  |Float        |The latitude of the stadium in decimal degrees
|longitude            |Longitude                 |Float        |The longitude of the stadium in decimal degrees
|opp                  |Opponent                  |String       |The opposing NFL team for this week
|result               |Game Result               |String       |The result of the game from the team's perspective
|points_for           |Points Scored             |Integer      |The number of points the team scored this week
|points_allowed       |Points Allowed            |Integer      |The number of points the team allowed the opposing team to score
|tot_yds              |Total Yards               |Integer      |The number of yards the team gained on offense
|pass_yds             |Passing Yards             |Integer      |The number of yards the team gained via passing
|rush_yds             |Rushing Yards             |Integer      |The number of yards the team gained via rushing
|first_downs          |First Downs               |Integer      |The number of first downs the team had
|turnovers            |Turnovers                 |Integer      |The number of turnovers lost
|opp_tot_yds          |Total Yards Allowed       |Integer      |The number of yards the team allowed on defense
|opp_pass_yds         |Passing Yards Allowed     |Integer      |The number of yards the team allowed via passing
|opp_rush_yds         |Rushing Yards Allowed     |Integer      |The number of yards the team allowed via rushing
|opp_first_downs      |First Downs Allowed       |Integer      |The number of first downs the opposing team had
|opp_turnovers        |Turnovers Gained          |Integer      |The number of turnovers the team gained on defense
|exp_pts_off          |Expected Points from Offense        |Float  |The expected points gained from offensive plays
|exp_pts_def          |Expected Points from Defense        |Float  |The expected points gained from defensive plays
|exp_pts_st           |Expected Points from Special Teams  |Float  |The expected points gained from special teams
|opp_index            |Opponent's Index          |String       |The index for the opponent's perspective of this game
|distance_travelled   |Distance Travelled        |Float        |Distance in miles the team travelled for the game
|distance_travelled_opp |Opponent's Distance Travelled |Float  |Distance in miles the opponent travelled for the game
|distance_travelled_opp_diff |Distance Travelled Difference |Float |How many more miles the team travelled than their opponents
|elo_start           |Starting Elo Score |Float |The Elo Score Rating of the team prior to the game
|elo_start_opp |Opponent's Starting Elo Score |Float |The Elo Score Rating of the opponent prior to the game
|elo_opp_diff_team |Elo Difference |Float |The difference in Elo Score before the game between the opponent and the team
|exp_win_prob |Expected Win Probability |Float |The probabilty the team has of winning the game, based on the Elo of the team and their opponents
|result_value |Result Value |Float |A number based on the outcome of the game to help calculate changes in Elo (1 for Win, 0.5 for Tie, 0 for Loss)
|mov_winner       |Winner's Margin of Victory |Float |The number of points the game's winner won by
|elo_winner_diff |Winner's Elo Difference |Float |The winner's starting Elo Score minus the loser's starting Elo score (negative values indicate upsets)
|mov_mult |Margin of Victory Multiplier |Float |A multiplier used to calculate changes in Elo score based on the margin of victory of the game
|elo_change |Change in Elo |Float |The value that a team's Elo score changes following this game
|elo_end |Ending Elo Score |Float |The value of a team's Elo score after the game




### Notes
 - Average fields calculate the average values for the team in that statistic over the previous x games, not including the current game (recognized by the 'avg' in the column name
 - Fields ending in '_opp' are those taken from the opponent's perspective of the same game (joining on Opponent's Index)
 - Rest Days for Week 1 games is set to 10
 - Values for Game Results are 'W' for win, 'L' for loss, 'T' for tie, and '-' for a cancelled game
 - Passing Yards and Passing Yards Allowed include yards lost from a sack
 - For information about expected points, please see Pro Football Reference's [blog post](https://www.sports-reference.com/blog/2012/03/features-expected-points/?__hstc=223721476.704f6f4ff227fa125ec62e7520a2aebf.1696284740229.1697921238423.1698078729540.9&__hssc=223721476.12.1698078729540&__hsfp=3421363042).

Let's take a look at the Elo scores of each team following the end of the 2022 regular season.

The top five teams all finished with at least 12 wins, winning at least 75% of their games that season. Elo does not represent winning percentage perfectly, but does give a sense of the success of the team at the point in the season. The Chiefs and Eagles had the best record at 14 and 3, and they are near the top of the list. The Steelers had the eighth highest Elo at the end of the season, but had a record of 9-8. Despite finishing just above .500, they ended the season with a high Elo because they won 7 of their last 9 games.

At the bottom of the list, we have five teams that each won five or fewer games. The order does seem to align closely with the team's records, and where they would wind up in the NFL draft. The Jets had a better record than some of the teams above them, but only winning one of their final eight games drops their Elo score.
